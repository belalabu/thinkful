```python
data = pd.read_csv('ESS.csv')
df = pd.DataFrame(data)
df.year.value_counts()
```

### 1) Did people become less trusting from 2012 to 2014? Compute results for each country in the sample.
Answer: Yes, from average value of 5.56 down to 5.55

```python
df.groupby('year')['ppltrst'].mean()
```

### 2) Did people become happier from 2012 to 2014? Compute results for each country in the sample.
Answer: On average people are less happy

See below for individual stats

```python
df1 = df[df.year == 6] #2012 data
df2 = df[df.year == 7] #2014 data

happy1 = df1.groupby('cntry')['happy'].mean()
happy2 = df2.groupby('cntry')['happy'].mean()
print(happy1, happy2)
```
