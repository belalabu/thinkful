```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
%matplotlib inline
import seaborn as sn

data = pd.read_csv('carbon.csv')
df = pd.DataFrame(data)

data1 = pd.read_csv('carbonweekly.csv')
df1 = pd.DataFrame(data1)

# find zscores
z = np.abs(stats.zscore(df1.iloc[:, 4]))
# get indices of outliers
outliers= list(np.where(z > 3)[0])
# drop them
df1= df1.drop(outliers)


figure(num=None, figsize=(14, 5), dpi=100, facecolor='w', edgecolor='k')
plt.plot( df['year'], df['mean'], linestyle='-', marker='o', color='green', alpha=.7)
plt.plot( df1['dymd'], df1['ppm'], linestyle='-', marker='', color= 'black', alpha =.4)
plt.ylim(300,420)
plt.xlabel('Year')
plt.ylabel('Parts Per Millon(ppm)')
plt.title('Mauna Loa Observatory Atmospheric Carbon Dioxide Readings')

plt.style.use('seaborn')

plt.show()
```

```python
figure(num=None, figsize=(10, 8), dpi=100, facecolor='w', edgecolor='k')
plt.bar(df['year'], df['mean'], width=0.6, color='mediumseagreen')
plt.ylim(300, 420)

plt.xlabel('Year')
plt.ylabel('Parts Per Millon(ppm)')
plt.title('Mauna Loa Observatory Atmospheric Carbon Dioxide Readings')

plt.style.use('seaborn')

plt.show()
```

```python
figure(num=None, figsize=(10, 8), dpi=100, facecolor='w', edgecolor='k')
plt.fill_between( df['year'], df['mean'], linestyle='-', color='green', alpha=.3)
plt.plot( df['year'], df['mean'], linestyle='-', color='green', alpha=.8)
plt.ylim(300,420)
plt.xlabel('Year')
plt.ylabel('Parts Per Millon(ppm)')
plt.title('Mauna Loa Observatory Atmospheric Carbon Dioxide Readings')

plt.style.use('seaborn')

plt.show()
```
