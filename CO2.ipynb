```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
%matplotlib inline
import seaborn as sn

data = pd.read_csv('carbon.csv')
df = pd.DataFrame(data)

figure(num=None, figsize=(10, 8), dpi=100, facecolor='w', edgecolor='k')
plt.plot( df['year'], df['mean'], linestyle='-', marker='o', color='green', alpha=.5)
plt.ylim(300,420)
plt.xlabel('Year')
plt.ylabel('Parts Per Millon(ppm)')
plt.title('Mauna Loa Observatory Atmospheric Carbon Dioxide Readings')

plt.style.use('seaborn')

plt.show()
```

```python
figure(num=None, figsize=(10, 8), dpi=100, facecolor='w', edgecolor='k')
plt.bar(df['year'], df['mean'], width=0.6, color='mediumseagreen')
plt.ylim(300, 420)

plt.xlabel('Year')
plt.ylabel('Parts Per Millon(ppm)')
plt.title('Mauna Loa Observatory Atmospheric Carbon Dioxide Readings')

plt.style.use('seaborn')

plt.show()
```

```python
figure(num=None, figsize=(10, 8), dpi=100, facecolor='w', edgecolor='k')
plt.fill_between( df['year'], df['mean'], linestyle='-', color='green', alpha=.3)
plt.plot( df['year'], df['mean'], linestyle='-', color='green', alpha=.8)
plt.ylim(300,420)
plt.xlabel('Year')
plt.ylabel('Parts Per Millon(ppm)')
plt.title('Mauna Loa Observatory Atmospheric Carbon Dioxide Readings')

plt.style.use('seaborn')

plt.show()
```
