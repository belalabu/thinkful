```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
%matplotlib inline
from scipy import stats

data = pd.read_csv('wellcome.csv')
df = pd.DataFrame(data)

#make names more manageable
df.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)':'cost'}, inplace=True)
df.rename(columns={'Journal title':'journal'}, inplace=True)
df.rename(columns={'Article title':'article'}, inplace=True)

#normalize all titles to lowercase
df['article'] = df['article'].str.lower()
df['journal']=df['journal'].str.lower()

#remove decimals
cost_split = df['cost'].str.split('.', expand=True)
df['cost']= cost_split[0]

#remove symbols from numbers
df['cost'] = df['cost'].str.replace('£', '')
df['cost'] = df['cost'].str.replace('$', '')

#remove commas from numbers
df['cost'] = df['cost'].str.replace(',', '')
#remove strange outliers
df=df[np.logical_not(df['cost'].isin(["999999", '0', 'NaN']))]

#Change column from string to numeric, this is the issue that was ruining everything.
#DO THIS FIRST NEXT TIME 
df.cost = pd.to_numeric(df.cost)


```
